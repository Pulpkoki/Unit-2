```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
file1_df = pd.read_csv("Data_Marketing_Customer_Analysis_Round3.csv")
file1_df
```

```python
numerical = file1_df.select_dtypes(include=np.number)
numerical
```

```python
categorical = file1_df.select_dtypes(include=object)
categorical
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
listColor = ['pink', 'orange', 'purple', 'red', 'green', 'brown', 'yellow', 'blue']


fig, ax = plt.subplots(4,2, figsize = (8,12))

for i in range(1,9):
    plt.subplot(4,2,i)
    sns.distplot(numerical.iloc[:,i-1], color = listColor[i-1])
    plt.title(f'{numerical.columns[i-1].upper()}')
fig.subplots_adjust(hspace=1.5)
```

```python
fig, ax = plt.subplots(4,2, figsize = (10,15))
for i in range(1,9):
    plt.subplot(4,2,i)
    plt.hist(numerical.iloc[:,i-1], color = '#398f29', edgecolor = 'white', bins = 50)
    plt.title(f'{numerical.columns[i-1].upper()}')
```

```python
#For the numerical variables, check the multicollinearity between the features. Please note that we will use the column total_claim_amount later as the target variable.
numerical.corr()
```

```python
fig = plt.figure(figsize=(8,6))
sns.heatmap(numerical.corr(), cmap='vlag', square=True, annot=True)
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(numerical['total_claim_amount'], numerical['monthly_premium_auto'], s=.5)
plt.title("Correlation")
plt.xlabel("Total claim amount")  
plt.ylabel("Monthly premium auto")

df3 = file1_df[['total_claim_amount', 'monthly_premium_auto']]
```

```python
fig = plt.figure(figsize=(8,6))
plt.scatter(numerical['customer_lifetime_value'], numerical['monthly_premium_auto'], s=.5)
plt.title("Correlation")
plt.xlabel("customer_lifetime_value")
plt.ylabel("Monthly premium auto")
```
